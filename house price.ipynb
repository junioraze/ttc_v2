```python
from ipykernel.pylab.backend_inline import flush_figures
cols = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt','GarageArea']
def out_plot(column):
    plt.gcf().clear()
    _ = plt.scatter(train[str(column)], train['SalePrice'], alpha = 0.8, s=80, c = 'r')
    _.figure.set_size_inches(12,5)
    plt.suptitle('Buscando por outliers em {}'.format(column), fontsize = 20)
    flush_figures()

interact(out_plot, column=cols);
```

```python
#Quantidade de valores nulos por coluna #train
nan_count = train.isnull().sum().sort_values(ascending=False)
nan_percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
nan_data_train = pd.concat([nan_count, nan_percent], axis = 1, keys =['Total', 'Percent'])
nan_data_train[nan_data_train['Total'] > 0]
```

```python
#Quantidade de valores nulos por coluna #test
nan_count = test.isnull().sum().sort_values(ascending=False)
nan_percent = (test.isnull().sum()/test.isnull().count()).sort_values(ascending=False)
nan_data_test = pd.concat([nan_count, nan_percent], axis = 1, keys =['Total', 'Percent'])
nan_data_test[nan_data_test['Total'] > 0]
```

```python
#Dropando colunas com valores nulos > 40%
drop_train = list(nan_data_train[nan_data_train['Percent'] > 0.4].index)
drop_test = list(nan_data_test[nan_data_test['Percent'] > 0.4].index)
print('Train features : {}\nTeste Features : {}'.format(drop_train,drop_test))
```

```python
#Dropando
train.drop(drop_train,1, inplace=True)
test.drop(drop_test,1, inplace=True)
```

```python
#Preenchendo valores nulos com média
target = train['SalePrice']
train_num = train[train.select_dtypes(exclude = ["object"]).columns]
train_num = train_num.drop("SalePrice", 1)
print("NAs presentes na base de treino : " + str(train_num.isnull().values.sum()))
train_num = train_num.fillna(train_num.median())
print("NAs restantes : " + str(train_num.isnull().values.sum()))
```

```python
#Devido a ter separado as colunas númericas, também separaremos as categoricas 
#para concatenar a frente em uma base nova de treino
train_cat = train[train.select_dtypes(include = ["object"]).columns]
```

```python
cat_columns = train_cat.isnull().sum().sort_values(ascending=False)
nan_cat_columns = list(cat_columns[cat_columns[:] > 0].index)
def values(column):
    return train_cat[column].value_counts()
    
interact(values, column=nan_cat_columns);
```

```python
#corrigindo nulos categoricos
for column in nan_cat_columns:
    valor_comum = train_cat[column].value_counts().index[0]
    train_cat[column] = train_cat[column].fillna(valor_comum)
    print('coluna {} corrigida.'.format(column))
    
print('-----------------\nValores nulos restantes : {}'.format(train_cat.isnull().values.sum()))
```

```python
# Check the skew of all numerical features
from scipy.stats import skew 
skewed_feats = train_num.apply(lambda x: skew(x)).sort_values(ascending=False)
print("\nSkew in numerical features: \n")
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness
```

```python
skewness = skewness[abs(skewness) > 0.75]
print("There are {} skewed numerical features to Box Cox transform".format(skewness.shape[0]))

from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    train_num[feat] = boxcox1p(train_num[feat], lam)
```

```python
#Re-validando número de colunas de cada tipo com a subtração de 5 nas categóricas. 
print('num : {}\ncat : {}'.format(len(train_num.columns), len(train_cat.columns)))
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_num,target,test_size = 0.33,random_state= 42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
n_folds = 5
from sklearn.metrics import make_scorer
from sklearn.model_selection import KFold
scorer = make_scorer(mean_squared_error,greater_is_better = False)
def rmse_CV_train(model):
    kf = KFold(n_folds,shuffle=True,random_state=41).get_n_splits(train_num.values)
    rmse = np.sqrt(-cross_val_score(model,X_train,y_train,scoring ="neg_mean_squared_error",cv=kf))
    return (rmse)
def rmse_CV_test(model):
    kf = KFold(n_folds,shuffle=True,random_state=41).get_n_splits(train_num.values)
    rmse = np.sqrt(-cross_val_score(model,X_test,y_test,scoring ="neg_mean_squared_error",cv=kf))
    return (rmse)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
test_pre = lr.predict(X_test)
train_pre = lr.predict(X_train)
print('rmse on train',rmse_CV_train(lr).mean())
print('rmse on train',rmse_CV_test(lr).mean())
```

```python
# Plot predictions - Real values
plt.figure(figsize=(12,12))
plt.scatter(train_pre, y_train, c = "blue",  label = "Training data")
plt.scatter(test_pre, y_test, c = "black",  label = "Validation data")
plt.title("Linear regression")
plt.xlabel("Predicted values")
plt.ylabel("Real values")
plt.legend(loc = "upper left")
plt.plot([10.5, 13.5], [10.5, 13.5], c = "red")
plt.show()
```
